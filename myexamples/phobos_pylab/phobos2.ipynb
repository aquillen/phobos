```python
from scipy.ndimage.filters import median_filter
#varpidot = prec_dphidt(tt,np.squeeze(varpi[0,:]),10)
dt = tt[1] - tt[0]

def prec_rate(vec):    
    mvec = median_filter(np.squeeze(vec),20)
    dphidt =np.diff(mvec)
    dphidt =np.append(dphidt,dphidt[np.size(tt)-2])
    dphidt/= dt
    dphidt = median_filter(dphidt,20)
    return dphidt

dphidt = prec_rate(varpi[0])
dphidt2 = prec_rate(maarr[1])
plt.plot(tt,dphidt,'b.',ms=1)
plt.plot(tt,dphidt2,'r.',ms=1)
plt.ylim((-1e-3,2e-3))
```
